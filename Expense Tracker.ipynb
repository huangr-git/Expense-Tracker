```python
import pandas as pd
import os
from datetime import date
import matplotlib.pyplot as plt #this is to create customized graphs
from IPython.display import display #a display function that lets us render items in a more visually appealing way

# Load expense data and creating a expense sheet csv
csv_file = "Expense Sheet.csv"
if os.path.exists(csv_file):
    data = pd.read_csv(csv_file)
else:
    data = pd.DataFrame(columns=["Date", "Category", "Amount"])

# Load savings goals data and creating a csv
savings_file = "Savings Goals.csv"
if os.path.exists(savings_file): #checks that this csv file pathway exists every run
    savings_data = pd.read_csv(savings_file)
else:
    savings_data = pd.DataFrame(columns=["Goal Name", "Target Amount", "Current Amount", "Target Date"])

categories = ["Grocery", "Household", "Transportation"]

# Main menu loop
#self reminder: \n means to move the cursor into the next line
while True:
    print("\n" + "="*50) # purely for customization to generate ==========
    print("EXPENSE & SAVINGS TRACKER MENU")
    print("="*50) #seperates the main title from the subtitles with =========
    print("EXPENSES:")
    print("1. Add Expense")
    print("2. Show All Expenses (Table)")
    print("3. Show Expense Summary Graphs")
    print("\nSAVINGS GOALS:")
    print("4. Add Savings Goal")
    print("5. Add Money to Savings Goal")
    print("6. View All Savings Goals")
    print("7. Show Savings Goals Progress Chart")
    print("\n0. Exit")
#print("="*50) is just a formatting feature to improve visualization. Ex: we can do "-"*30 to create ------ but shorter
    print("="*50)
    opt = input("Choose option: ")
    
    # Add Expense (original functionality)
    if opt == '1':
        print('\nSelect an expense category:')
        for idx, cat in enumerate(categories, 1):
            print(f"{idx}. {cat}")
        item_choice = input('Choose a category number: ')
        try:
            item_idx = int(item_choice) - 1
            category = categories[item_idx]
        except (ValueError, IndexError):
            print('Invalid choice. Please Try again.')
            continue
        try:
            amount = float(input('Enter amount spent: '))
        except ValueError:
            print('Amount must be a number.')
            continue
        
        today = date.today().isoformat()
        new = pd.DataFrame([[today, category, amount]], columns=data.columns)
        data = pd.concat([data, new], ignore_index=True)
        data.to_csv(csv_file, index=False)
        print(f' Expense added: {category} – ${amount:.2f}!')
    
    # Show All Expenses (original functionality)
    elif opt == '2':
        if data.empty:
            print("\nNo expenses recorded yet.")
        else:
            print("\nYour Expenses Table:")
            display(data)
    
    # Show Expense Graphs (original functionality)
    elif opt == '3':
        if data.empty:
            print("No data to plot yet.")
            continue
        # Category breakdown Graph
        summary = data.groupby('Category')["Amount"].sum()
        summary.plot(kind='bar', color='skyblue', title='Total Spent by Category')
        plt.ylabel('Amount')
        plt.show()
        # Daily trend graph
        date_summary = data.groupby('Date')["Amount"].sum()
        date_summary.plot(kind='line', marker='o', color='green', title='Total Spent per Day')
        plt.ylabel('Amount')
        plt.show()
    
    # Add Savings Goal
    elif opt == '4':
        print("\n--- Add A New Savings Goal ---")
        goal_name = input("Enter goal name (ex: Emergency Fund, Vacation): ")
        try:
            target_amount = float(input("Enter target amount ($): "))
            target_date = input("Enter target date (DD/MM/YYYY): ")
            
            new_goal = pd.DataFrame([[goal_name, target_amount, 0.0, target_date]], 
                                   columns=savings_data.columns)
            savings_data = pd.concat([savings_data, new_goal], ignore_index=True)
            savings_data.to_csv(savings_file, index=False)
            print(f"Savings goal '{goal_name}' added successfully!")
        except ValueError:
            print("Invalid input. Please try again.")
    
    # Add Money to Savings Goal
    elif opt == '5':
        if savings_data.empty:
            print("\nNo savings goals yet. Please add a goal first (option 4).")
            continue
        
        print("\nYour Savings Goals:")
        for idx, row in savings_data.iterrows():
            print(f"{idx + 1}. {row['Goal Name']} (Current: ${row['Current Amount']:.2f} / Target: ${row['Target Amount']:.2f})")
        
        try:
            goal_choice = int(input("\nChoose a goal number: ")) - 1
            if goal_choice < 0 or goal_choice >= len(savings_data):
                print("Invalid goal number.")
                continue
            
            amount = float(input("Enter amount to add ($): "))
            savings_data.at[goal_choice, 'Current Amount'] += amount
            savings_data.to_csv(savings_file, index=False)
            
            goal_name = savings_data.at[goal_choice, 'Goal Name']
            current = savings_data.at[goal_choice, 'Current Amount']
            target = savings_data.at[goal_choice, 'Target Amount']
            
            print(f" Added ${amount:.2f} to '{goal_name}'")
            print(f"Progress: ${current:.2f} / ${target:.2f} ({(current/target*100):.1f}%)")
            
            if current >= target:
                print(f"Congratulations! You've reached your '{goal_name}' goal!")
        except (ValueError, IndexError):
            print("Invalid input. Please try again.")
    
    # View All Savings Goals
    elif opt == '6':
        if savings_data.empty:
            print("\nNo savings goals yet. Add a goal first (option 4).")
            continue
        
        print("\n" + "="*80)
        print("YOUR SAVINGS GOALS")
        print("="*80)
        
        
#UNDER HERE CONSISTS OF DISPLAY OF SAVING/EXPENSES AND GRAPHS/CHARTS        
        # Create display dataframe with calculations
        display_savings = savings_data.copy()
        display_savings['Progress (%)'] = (display_savings['Current Amount'] / 
                                           display_savings['Target Amount'] * 100).round(1)
        display_savings['Remaining'] = display_savings['Target Amount'] - display_savings['Current Amount']
        
        # Format currency columns
        for col in ['Target Amount', 'Current Amount', 'Remaining']:
            display_savings[col] = display_savings[col].apply(lambda x: f"${x:,.2f}")
        display_savings['Progress (%)'] = display_savings['Progress (%)'].apply(lambda x: f"{x}%")
        
        display(display_savings)
        print("="*80)
    
    # Show Savings Goals Progress Chart
    elif opt == '7':
        if savings_data.empty:
            print("\nNo savings goals to plot yet.")
            continue
        
        # Calculate progress percentages
        progress = (savings_data['Current Amount'] / savings_data['Target Amount'] * 100).clip(upper=100)
        goals = savings_data['Goal Name']
        
        # Create color coding based on progress of savings account
        colors = []
        for p in progress:
            if p >= 100:
                colors.append('green')
            elif p >= 75:
                colors.append('orange')
            elif p >= 50:
                colors.append('gold')
            else:
                colors.append('red')
#Customizing the graph's labels, titles, and colors        
        plt.figure(figsize=(10, 6))
        plt.barh(goals, progress, color=colors, alpha=0.7, edgecolor='black')
        plt.xlabel('Progress (%)', fontsize=12, fontweight='bold')
        plt.ylabel('Savings Goals', fontsize=12, fontweight='bold')
        plt.title('Savings Goals Progress', fontsize=14, fontweight='bold')
        plt.xlim(0, 100)
        plt.axvline(x=100, color='green', linestyle='--', linewidth=2, alpha=0.5, label='Target')
        
        # Add % labels on bars
        for i, (goal, prog) in enumerate(zip(goals, progress)):
            plt.text(prog + 2, i, f'{prog:.1f}%', va='center', fontsize=10, fontweight='bold')
        
        plt.legend()
        plt.tight_layout()
        plt.show()
    
    # Exit
    elif opt == '0':
        data.to_csv(csv_file, index=False)
        savings_data.to_csv(savings_file, index=False)
        print('\n Thank you for using Expense & Savings Tracker! All Data Has Been Saved.')
        break
    
    else:
        print('Invalid option. Try again.')
  
#Overall, what's cool about it is that you can run this code multiple times, even like every day or other day
#And it will automatically pull up the old expenses that you have listed in the saved csv file.
#The daily trend graph I added is a line graph that lets you see the increase or decrease trend of you spending overtime
```
